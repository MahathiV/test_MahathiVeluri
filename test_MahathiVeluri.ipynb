```python
#import libraries

import  os
import numpy as np
import pandas as pd
```

```python
#reading first csv file from folder

csv1 = os.path.join("Resources","StudentsPerformance - StudentsPerformance.csv")
```

```python
#csv to dataframe
student_df = pd.read_csv(csv1)
student_df.head()
```

```python
#Q1: Which 'parent level of education' is highest for any 'race/ethnicity'? 
#Output should print 'parent level of education' and count (eg. group c 65)
```

```python
# Grouping by 'parent level of education' and 'race/ethinicity' to compute the number of occurrences
# Value calculated into a new df and max is desplayed accordingly

count_df = student_df[['parental level of education','race/ethnicity', 'gender']].groupby(['parental level of education','race/ethnicity']).count()
count_df.rename(columns={'gender': 'count'}, inplace=True)
count_df.reset_index(inplace=True)
count_df.max()
```

```python
#Q2: Which 'gender' has more occurrence where 'math score' greater than 'reading score' but less than 'writing score'? 
#Output should print 'parent level of education' and count 
#(eg. female occurrence_mathsscore occurrence_readingscore occurrence_writingscore)
```

```python
# DF is filtered to compute the cells with the given condition
# Filtered DF is grouped by the gender to compute the count of each gender

math_score = student_df['math score']
reading_score = student_df['reading score']
writing_score = student_df['writing score']
filter_df = student_df[(math_score > reading_score) & (math_score < writing_score)]
filter_df.reset_index(drop=True, inplace=True)
filter_df.groupby(['gender']).count()
```

```python
#Q3: Display 'parental level of education' and total score for students who took standard lunch and complete the course? 
#(Output should all fields and relevant values)
```

```python
# DF is filtered to compute the cells with the given condition
# Total score is computed in the Filtered DF by adding the math, reading and writing scores, and output is printed

score_df = student_df[(student_df['lunch']=="standard") & (student_df['test preparation course'] == "completed" )]
score_filter = score_df[['parental level of education','math score','reading score','writing score']]
score_filter.reset_index(drop=True, inplace=True)

score_filter['Total_Score'] = score_filter['math score'] + score_filter['reading score'] + score_filter['writing score']
score_filter[['parental level of education','Total_Score']]
```

```python
#Q4: Display median value for total score for female students who couldn't complete the course 
#(Output should all fields and relevant values)
```

```python
# DF is filtered to compute the cells with the given condition
# Total score is computed in the Filtered DF by adding the math, reading and writing scores
# Median is computed by using pandas median function

female_df = student_df[(student_df['gender']=="female") & (student_df['test preparation course'] == "none" )]
female_df['Total_Score'] = female_df['math score'] + female_df['reading score'] + female_df['writing score']
female_df['Total_Score'].median()
```

```python
#Q5: Consider different combinations among values of - 
#'gender', 'race/ethnicity', 'parental level of education','lunch','test' and 'preparation course'. 
#Print the combination where average math score is greater than average reading score but less than average writing score
```

```python
student_df = pd.read_csv(csv1)
col_list = [i for i in list(student_df.columns) if i not in ["math score", "reading score", "writing score"]]
student_df['combined'] = student_df[col_list].apply(lambda row: '_'.join(row.values.astype(str)), axis=1)
comb_df = student_df[["combined", "math score", "reading score", "writing score"]]

avg_df_1 = pd.DataFrame()
avg_df_1['average math score'] = comb_df.groupby('combined')['math score'].mean()
avg_df_1.reset_index(inplace=True)

avg_df_2 = pd.DataFrame()
avg_df_2['average reading score'] = comb_df.groupby('combined')['reading score'].mean()
avg_df_2.reset_index(inplace=True)

avg_df_3 = pd.DataFrame()
avg_df_3['average writing score'] = comb_df.groupby('combined')['writing score'].mean()
avg_df_3.reset_index(inplace=True)

merged_df = avg_df_1.merge(avg_df_2, how='left', left_on=['combined'], right_on=['combined']).merge(avg_df_3, how='left', left_on=['combined'], right_on=['combined'])
out_df = merged_df[(merged_df['average math score'] > merged_df['average reading score']) & (merged_df['average math score'] < merged_df['average writing score'])]
list(out_df['combined'])
```

```python
#Q10: Telco company wants to analyse pattern of customers who churned. 
#Can you help them know which combination of all the fields is most risky and must be avoided?
```

```python
# Take all columns other than customerID, MonthlyCharges, TotalCharges and churn --> l1
# Create 2 new columns monthly_charges_bucket and total_charges_bucket --> l2
# Take l1 + l2 for analysis
# Find all unique combinations... For each combination, compute (churned / Total) and calculate the ones with high 
# value of this ratio
tele_df = pd.read_csv(csv2)
# print(len(tele_df))
col_list = [i for i in list(tele_df.columns) if i not in ["customerID", "MonthlyCharges", "TotalCharge", "Churn"]]
tele_df['monthly_charges_bucket'] = pd.cut(tele_df['MonthlyCharges'], 10)
tele_df['total_charges_bucket'] = pd.cut(tele_df['TotalCharges'], 100)
col_list += ['monthly_charges_bucket', 'total_charges_bucket']
tele_df['combined'] = tele_df[col_list].apply(lambda row: '_'.join(row.values.astype(str)), axis=1)
df11=tele_df.groupby('combined')['Churn'].apply(lambda x: (x=='Yes').sum()).reset_index(name='churn_count')
df12=tele_df[['combined', 'Churn']].groupby('combined').agg('count')
df12.reset_index(inplace=True)
df12.rename(columns={'Churn': 'total_count'}, inplace=True)
df_merged = df11.merge(df12, how='left', left_on=['combined'], right_on=['combined'])
df_merged['churn_fraction'] = df_merged['churn_count'] / df_merged['total_count']

# set(df_merged['churn_fraction']) --> Output is: {0.0, 0.25, 0.3333333333333333, 0.5, 0.6666666666666666, 1.0}

list(df_merged[df_merged['churn_fraction']==1.0]['combined'])
```
