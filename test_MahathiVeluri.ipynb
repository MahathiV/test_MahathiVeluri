```python
#Q3: Display 'parental level of education' and total score for students who took standard lunch and complete the course? 
#(Output should all fields and relevant values)
```

```python
score_df = student_df.loc[(student_df['lunch']=="standard") & (student_df['test preparation course'] == "completed" )]
```

```python
score_filter = score_df[['parental level of education','math score','reading score','writing score']]
```

```python
score_filter['Total_Score'] = score_filter ['math score'] + score_filter['reading score'] + score_filter['writing score']
```

```python
score_filter[['parental level of education','Total_Score']]
```
